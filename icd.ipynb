```python
import pandas as pd

# タスク×スキル対応表.csvのパスを挿入
task_skill = pd.read_csv('', sep=',', encoding='shift-JIS')
task_skill.head()
```

```python
# タスク一覧.csvのパスを挿入
tasks = pd.read_csv('', sep=',', encoding='shift-JIS')
tasks.head()
```

```python
# スキル一覧（スキル項目）.csvのパスを挿入
skills = pd.read_csv('', sep=',', encoding='shift-JIS')
skills.head()
```

```python
# 評価項目コード→タスク小分類コード(1対1なので1つ)
evaluation_code = input('評価項目コード：')
print('タスク小分類コード：')
print(tasks[tasks['評価項目コード']==evaluation_code].loc[:, 'タスク小分類コード'][0])
```

```python
# 評価項目コード→スキル項目コード(1対多なので複数)
evaluation_code = input('評価項目コード：')
print('スキル項目コード：')
task_s_category_code = tasks[tasks['評価項目コード']==evaluation_code].loc[:, 'タスク小分類コード']
display(task_skill[task_skill['タスク小分類コード']==task_s_category_code[0]].loc[:, 'スキル項目コード'])
```

```python
# 評価項目コード→その学習者のスキル詳細(1対多なので複数)

task = input('該当箇所のリンク：')
# task = 'https://github.com/HazeyamaLab/SE00G0/blob/'  # 確認用
evaluation_code = input('評価項目コード：')
# evaluation_code = 'ST01.1.1.1'  # 確認用

task_s_category_code = tasks[tasks['評価項目コード']==evaluation_code].loc[:, 'タスク小分類コード']
skill_codes = task_skill[task_skill['タスク小分類コード']==task_s_category_code[0]].loc[:, 'スキル項目コード']
task_s_category = tasks[tasks['評価項目コード']==evaluation_code].loc[:, 'タスク小分類']
evaluation_items = tasks[tasks['評価項目コード']==evaluation_code].loc[:, '評価項目']
for skill_code in skill_codes:
  skill_items = skills[skills['スキル項目コード']==skill_code].loc[:, 'スキル項目']
  for skill_item in skill_items:
    print('#', skill_item)
    for i in task_s_category:
      print('##', i)
      for i in evaluation_items:
        print('以下のコードから，' + i + 'ことができていると判断しました．')
        print(task)
        print('')
```
